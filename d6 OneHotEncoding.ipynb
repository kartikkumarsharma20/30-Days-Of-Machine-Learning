```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv("Salary_Dataset.csv")
df.head()
```

```python
en_data=df[["country","Purchased"]]
Ohe=OneHotEncoder()
en_arr=Ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(en_arr,columns=["Purchased_No","Purchased_Yes","country_Australia","country_Canada","country_Dubai","country_USA"])
```

```python
en_purchase_data=df[["Purchased"]]
OHE=OneHotEncoder(drop="first")
arr=OHE.fit_transform(en_purchase_data).toarray()
arr
```

```python
pd.DataFrame(arr,columns=["Purchased_Yes"]).head()
```

```python
df.drop(columns="Purchased",inplace=True)
df["Purchased_Yes"]=pd.DataFrame(arr,columns=["Purchased_Yes"])
df.head()
```
