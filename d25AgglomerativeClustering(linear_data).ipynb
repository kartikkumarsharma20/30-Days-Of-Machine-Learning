```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
```

```python
df=pd.read_csv("Iris.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
import scipy.cluster.hierarchy as sp
```

```python
plt.figure(figsize=(20,15))
sp.dendrogram(sp.linkage(df, method='single', metric='euclidean'))
plt.show()
```

```python
agc=AgglomerativeClustering(n_clusters=2,linkage="single")
df["predict"]=agc.fit_predict(df)
```

```python
sns.pairplot(df,hue="predict")
plt.show()
```
