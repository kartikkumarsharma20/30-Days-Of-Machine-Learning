```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
x,y=make_moons(n_samples=500,noise=0.05)
data={"data1":x[:,0],"data2":x[:,1]}
df=pd.DataFrame(data)
y
```

```python
df.head()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.scatter(df["data1"],df["data2"])
plt.show()
```

```python
db=DBSCAN(eps=0.3,min_samples=6)
df["predict"]=db.fit_predict(df)
df["predict"]
```

```python
sns.pairplot(data=df,hue="predict")
plt.show()
```
